### 4- drop all the useless columns and  rows with too much NaN ,
> tip : useless columns can be like `Internal ID`and can be also columns with too much NaN
- rename columns for easy access

```python
# dropping useless columns
df.drop(columns=["Internal ID", "Click Coordinates (x, y)", "Unnamed: 113"], inplace=True)
```

```python
def delete_semi_Empty_cols(df):
    half_empty_cols = []
    for col in df.columns:
        if df[col].isnull().sum()/df.shape[0] > 0.35:
            half_empty_cols.append(col)
    df.drop(columns=half_empty_cols, inplace=True)
    return df

delete_semi_Empty_cols(df)
df.shape
```

### 5 - clean `going out` column
> tip : you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
df = df.dropna(subset=["Q5: STATE, PROVINCE, COUNTY, ETC", "Q4: COUNTRY", "Q1: GOING OUT?", "Q3: AGE"])
```

### 6 - clean ` gender` column
> tip : you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
df["Q2: GENDER"] = df["Q2: GENDER"].replace({'Male':0, 'Female':1, "I'd rather not say": 3, 'Other':4}).fillna(-1)
pd.to_numeric(df["Q2: GENDER"])
```

### 7 - clean ` age` column
> tip : note the type of age column , and NaN values
 - try to use `to_numeric`
 - tip : you should choose which best value to put in place of these NaNs ( mean, median, ....etc)

```python
from sklearn.impute import SimpleImputer
df["Q3: AGE"] = pd.to_numeric(df["Q3: AGE"], errors="coerce")

imputer = SimpleImputer(strategy="mean")
df["Q3: AGE"] = imputer.fit_transform(df[["Q3: AGE"]])
pd.to_numeric(df["Q3: AGE"])
```

### 8 - clean ` country` column
> tip : note that country column has country names with different writing ways
  - use fuzzywuzzy to deal with this problem
  - you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
from fuzzywuzzy import fuzz
country_list = ["United States", "Afghanistan", "Albania", "Algeria", "Andorra", "Angola", "Argentina", "Armenia", 
                "Australia", "Austria", "Brazil", "Canada", "China", "Egypt", "France", "Germany", "India", "Italy", 
                "Japan", "United Kingdom", "Saudi Arabia", "South Africa"]


abbreviation_map = {
    "us": "United States",
    "usa": "United States",
    "u.s.a": "United States",
    "uk": "United Kingdom",
    "england": "United Kingdom",
    "brasil": "Brazil",
    "uae": "United Arab Emirates",
    "korea": "South Korea"
}

def process_country(country_name):
    if pd.isna(country_name): 
        return None
    country_name.strip().lower()

    if country_name in abbreviation_map:
        return abbreviation_map[country_name]

    best_match = max(
        [(country, fuzz.partial_ratio(country_name, country.lower())) for country in country_list],
        key=lambda x: x[1]
    )
    
    return best_match
df["Q4: COUNTRY"] = df["Q4: COUNTRY"].apply(process_country)
```

### 9 - clean ` area` column
> tip : area column has the same problem as country columns , but looking at all it's unique values may give us another way to deal with it rather than try to fix it

```python
state_list = [
    "Alabama", "Alaska", "Arizona", "Arkansas", "California", "Colorado", "Connecticut", "Delaware", "Florida", 
    "Georgia", "Hawaii", "Idaho", "Illinois", "Indiana", "Iowa", "Kansas", "Kentucky", "Louisiana", "Maine", 
    "Maryland", "Massachusetts", "Michigan", "Minnesota", "Mississippi", "Missouri", "Montana", "Nebraska", 
    "Nevada", "New Hampshire", "New Jersey", "New Mexico", "New York", "North Carolina", "North Dakota", 
    "Ohio", "Oklahoma", "Oregon", "Pennsylvania", "Rhode Island", "South Carolina", "South Dakota", 
    "Tennessee", "Texas", "Utah", "Vermont", "Virginia", "Washington", "West Virginia", "Wisconsin", "Wyoming", 
    "District of Columbia", "British Columbia", "Ontario", "Quebec", "Alberta", "Nova Scotia", "Manitoba",
    "Newfoundland and Labrador", "Prince Edward Island", "Saskatchewan", "New Brunswick"
]

state_map = {
    "ca": "California", "ny": "New York", "tx": "Texas", "co": "Colorado", "fl": "Florida", "nj": "New Jersey",
    "pa": "Pennsylvania", "nc": "North Carolina", "wa": "Washington", "dc": "District of Columbia", 
    "ontario": "Ontario", "bc": "British Columbia", "ab": "Alberta", "qc": "Quebec", "ri": "Rhode Island",
    "il": "Illinois", "mi": "Michigan", "oh": "Ohio", "va": "Virginia", "mo": "Missouri", "mn": "Minnesota",
    "ma": "Massachusetts", "in": "Indiana", "az": "Arizona", "or": "Oregon", "ga": "Georgia", "md": "Maryland",
    "wi": "Wisconsin", "la": "Louisiana", "nh": "New Hampshire", "vt": "Vermont", "tn": "Tennessee",
    "nv": "Nevada", "mt": "Montana", "ky": "Kentucky", "ar": "Arkansas", "id": "Idaho", "ia": "Iowa",
    "ms": "Mississippi", "nd": "North Dakota", "sd": "South Dakota", "me": "Maine", "ak": "Alaska",
    "al": "Alabama", "sc": "South Carolina", "nm": "New Mexico", "hi": "Hawaii", "ks": "Kansas", "ok": "Oklahoma",
    "ne": "Nebraska", "wy": "Wyoming", "wv": "West Virginia"
}

def process_state(state_name):

    state_name_normalized = state_name.strip().lower()

    if state_name_normalized in state_map:
        return state_map[state_name_normalized]

    best_match = max(
        [(state, fuzz.partial_ratio(state_name_normalized, state.lower())) for state in state_list],
        key=lambda x: x[1]
    )
    return best_match[0]

df["Q5: STATE, PROVINCE, COUNTY, ETC"] = df["Q5: STATE, PROVINCE, COUNTY, ETC"].apply(process_state)

```

### 10 - clean ` Q6` columns
>- tip : Q6 has all kinds of chocolate bars and questions with three distinct options to choose from (Joy, Meh, Despair).
- so all the column that has Q6 is a candy name , so you should go through all these column and you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
for col in df.columns:
    if "Q6" in col:
        df[col] = df[col].replace({'JOY': 2, 'MEH': 1, 'DESPAIR': 0}).fillna(-1)
        pd.to_numeric(df[col])
```

### 11 - clean ` dress` column
>tip : you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
df["Q10: DRESS"] = df["Q10: DRESS"].replace({'White and gold': 0, "Blue and black" : 1}).fillna(-1)
pd.to_numeric(df["Q10: DRESS"])
```

### 12 - clean ` day` column
>tip : you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
df["Q11: DAY"] = df["Q11: DAY"].replace({'Friday': 0, "Sunday" : 1}).fillna(-1)
pd.to_numeric(df["Q11: DAY"])
```
